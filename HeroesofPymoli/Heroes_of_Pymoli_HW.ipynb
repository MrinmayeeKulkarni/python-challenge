```python
import pandas as pd
import numpy as np
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
player_demographics=purchase_data.loc[:,('SN','Age','Gender')]
player_dem=player_demographics.drop_duplicates()
playercount=len(purchase_data["SN"].unique())
playerscounted=pd.DataFrame({"Total Number of Players": [playercount]})
playerscounted
```

```python
unique_items=purchase_data["Item ID"].nunique()
avg_price=round(purchase_data["Price"].mean(),2)
numberofpurchases=len(purchase_data["Purchase ID"])
totalrevenue=round(purchase_data["Price"].sum(),2)
```

```python
summary_df=pd.DataFrame({"Number of Unique Items":[unique_items],
                         "Average Price":[avg_price],
                         "Number of Purchases":[numberofpurchases],
                         "Total Revenue": "$"+str(totalrevenue)})
summary_df
```

```python
totaldemcount=purchase_data["Gender"].value_counts()
pct=round((totaldemcount/len(purchase_data["Gender"]))*100, 2)
```

```python
summary=pd.concat([totaldemcount,pct],axis=1,keys=('Total Count','Percentage of Count'))
summary.head()
```

```python
purchasing_analysis=purchase_data.groupby('Gender')
countbygender=player_dem["Gender"].value_counts()
summarypa=round((purchasing_analysis["Price"].mean()),2)
total=round((purchasing_analysis["Price"].sum()),2)
avgtotal=round((total/countbygender),2)
```

```python
purchanalysis_table=pd.DataFrame({"Total Count":totaldemcount,
                                 "Average purchase price":summarypa,
                                 "Total Purchase value":total,
                                "Average total purchase":avgtotal
                                 })
purchanalysis_table["Average purchase price"]=purchanalysis_table["Average purchase price"].map("${:.2f}".format)
purchanalysis_table["Total Purchase value"]=purchanalysis_table["Total Purchase value"].map("${:.2f}".format)
purchanalysis_table["Average total purchase"]=purchanalysis_table["Average total purchase"].map("${:.2f}".format)
purchanalysis_table
```

```python
bins=[0,9,14,19,24,29,34,39,45]
group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
player_dem["Age groups"]=pd.cut(player_dem["Age"],bins,labels=group_labels)
Agedemo=player_dem.groupby("Age groups")
agecount=Agedemo["SN"].count()
pctplayers=round(((agecount/playercount)*100),2)
```

```python
summaryagedemo=pd.DataFrame({"Total count":agecount,
                            "Percentage of Players":pctplayers})
summaryagedemo
```

```python
bins=[0,9,14,19,24,29,34,39,45]
group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age groups"]=pd.cut(purchase_data["Age"],bins,labels=group_labels)
Age_analysis=purchase_data.groupby("Age groups")
avgprice_age=round((Age_analysis["Price"].mean()),2)
totalvalue_age=round((Age_analysis["Price"].sum()),2)
avgtotal_age=round((totalvalue_age/agecount),2)
```

```python
summary_purchasinganalysis_age=pd.DataFrame({"Total count":agecount,
                                            "Average Purchase Price":avgprice_age,
                                            "Total Purchase Value":totalvalue_age,
                                            "Avg Total Purchase per person":avgtotal_age})
summary_purchasinganalysis_age["Average Purchase Price"]=summary_purchasinganalysis_age["Average Purchase Price"].map("${:.2f}".format)
summary_purchasinganalysis_age["Total Purchase Value"]=summary_purchasinganalysis_age["Total Purchase Value"].map("${:.2f}".format)
summary_purchasinganalysis_age["Avg Total Purchase per person"]=summary_purchasinganalysis_age["Avg Total Purchase per person"].map("${:.2f}".format)
summary_purchasinganalysis_age
```

```python
topspenders=purchase_data.groupby("SN")
purchasecount=topspenders["SN"].count()
avgpurchprice=round((topspenders["Price"].mean()),2)
totalpurchprice=round((topspenders["Price"].sum()),2)
```

```python
spenders=pd.DataFrame({"Purchase Count":purchasecount,
                      "Average Purchase Price":avgpurchprice,
                      "Total Purchase Price":totalpurchprice})
top_spenders=spenders.sort_values("Total Purchase Price",ascending=False)
top_spenders["Average Purchase Price"]=top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Price"]=top_spenders["Total Purchase Price"].map("${:.2f}".format)
top_spenders.head()
```

```python
pop=purchase_data[["Item ID","Item Name","Price"]]
popular=pop.groupby(["Item ID","Item Name"])
purchasecount_popular=popular["Item Name"].count()
totalprice_popular=round((popular["Price"].sum()),2)
price_items=round((popular["Price"].mean()),2)
```

```python
popular_items=pd.DataFrame({"Purchase Count":purchasecount_popular,
                            "Item Price":price_items,
                            "Total Purchase Value":totalprice_popular})
mostpopular=popular_items.sort_values("Purchase Count",ascending=False)
mostpopular["Item Price"]=mostpopular["Item Price"].map("${:.2f}".format)
mostpopular["Total Purchase Value"]=mostpopular["Total Purchase Value"].map("${:.2f}".format)
mostpopular.head()
```

```python
mostprofitableitems=popular_items.sort_values("Total Purchase Value",ascending=False)
mostprofitableitems["Item Price"]=mostprofitableitems["Item Price"].map("${:.2f}".format)
mostprofitableitems["Total Purchase Value"]=mostprofitableitems["Total Purchase Value"].map("${:.2f}".format)
mostprofitableitems.head()
```
